```python
# Preparando e treinando o modelo
```

```python
investimentos.columns
```

```python
x = investimentos[['youtube', 'facebook', 'newspaper']]
y = investimentos[['sales']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7 , test_size = 0.3, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r = metrics.r2_score(y_test, y_pred)
```

```python
metrics.r2_score(y_true=y_test, y_pred=y_pred)
```
